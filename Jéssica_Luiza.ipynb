```python
Atividade ‚Äì Implementa√ß√£o do M√©todo de Jacobi em Python
Disciplina: C√°lculo Num√©rico
Enunciado
Voc√™ dever√° implementar, em Python, o m√©todo de Jacobi para resolver sistemas lineares do tipo ùê¥ùë•=ùëè.

Parte 1 ‚Äì Fun√ß√£o do m√©todo
Crie uma fun√ß√£o em Python que:

receba a matriz ùê¥ , o vetor C , um chute inicial, a toler√¢ncia e o n√∫mero m√°ximo de itera√ß√µes;

calcule as itera√ß√µes do m√©todo de Jacobi;

retorne:

a solu√ß√£o aproximada,o n√∫mero de itera√ß√µes realizadas,e mostre na tela, passo a passo, cada itera√ß√£o (valores de ùë• e erro).

Parte 2 ‚Äì Teste do algoritmo
Na parte principal do programa (fora da fun√ß√£o):

Defina pelo menos dois sistemas lineares diferentes (matrizes ùê¥ e vetores ùëè );

Para cada sistema, chame a fun√ß√£o criada e exiba o resultado final.

Requisitos
O c√≥digo deve estar organizado com a fun√ß√£o em um bloco e as chamadas da fun√ß√£o em outro.

Em cada itera√ß√£o, mostrar algo como:

Itera√ß√£o 1: x = [ ... ] erro = ...

Itera√ß√£o 2: x = [ ... ] erro = ...

ou uma tabela com os resultados de cada intera√ß√£o
```

```python
# FUN√á√ÉO DO M√âTODO

import numpy as np

def metodo_jacobi(J, d, chute, precisao, limite):
    J = J.astype(float)
    d = d.astype(float)
    vetor = chute.astype(float)
    tamanho = len(d)

    registro = []
    err_inicial = np.max(np.abs(vetor))
    registro.append((0, vetor.copy(), err_inicial))

    for passo in range(1, limite + 1):
        prox = np.zeros_like(vetor)

        for linha in range(tamanho):
            acumulador = sum(J[linha, col] * vetor[col] for col in range(tamanho) if col != linha)
            prox[linha] = (d[linha] - acumulador) / J[linha, linha]

        erro_atual = np.max(np.abs(prox - vetor))
        registro.append((passo, prox.copy(), erro_atual))

        if erro_atual < precisao:
            return prox, passo, registro

        vetor = prox

    return vetor, limite, registro
```

```python
# TESTE DO ALGORITMO

import numpy as np

def exibir_resultados(J, d, resposta, relatorio):
    print("M√©todo de Jacobi")
    print("Matriz Ampliada")

    for linha_idx in range(len(d)):
        texto = " ".join(f"{J[linha_idx, coluna]:7.4f}" for coluna in range(len(d)))
        print(f"{texto} {d[linha_idx]:7.4f}")

    print("\n k  COLUNA1   COLUNA2  COLUNA3  Erro")

    for passo, vet, falha in relatorio:
        print(f"{passo:2d}  {vet[0]:7.4f}  {vet[1]:7.4f}  {vet[2]:7.4f}  {falha:7.4f}")

    print("\nSolu√ß√£o do sistema")
    for item in resposta:
        print(f"   {item:.4f}", end="")
    print("\n")
```

```python
# TESTE

import numpy as np

J1 = np.array([
    [10, -2, 1],
    [-2, 8, -2],
    [1, -2, 6]
], dtype=float)

d1 = np.array([9, 5, 7], dtype=float)

chute_inicial = np.zeros(3)

res, passos, rel = metodo_jacobi(J1, d1, chute_inicial, precisao=1e-4, limite=50)

exibir_resultados(J1, d1, res, rel)
```
